```python
# Input root path.
ROOT_PATH = './'

# Learning parameters.
BATCH_SIZE = 32
LR = 0.001
EPOCHS = 400
IMG_ORIG_SIZE = 224
IMGAE_RESIZE = 96
MONITOR_VAL = True
NUM_KEYPOINTS = 32

# Train/test split.
TEST_SPLIT = 0.15
VAL_SPLIT = 0.15

# Show dataset keypoint plot (executing `dataset.py`).
SHOW_DATASET_PLOT = True

SEED = 123
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('ggplot')
def evaluation_keypoints_plot(
    image, outputs, orig_keypoints, save_path
):
    """
    This function plots the regressed (predicted) keypoints from all the 
    evalutaion images.
    """
    output_keypoint = outputs.reshape(-1, 2)
    orig_keypoint = orig_keypoints.reshape(-1, 2)
    image = image.reshape(IMG_ORIG_SIZE, IMG_ORIG_SIZE, 3)
    plt.style.use('default')
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    for p in range(output_keypoint.shape[0]):
        plt.plot(output_keypoint[p, 0], output_keypoint[p, 1], 'r.')
        plt.text(output_keypoint[p, 0], output_keypoint[p, 1], f"{p}")
        plt.plot(orig_keypoint[p, 0], orig_keypoint[p, 1], 'g.')
        plt.text(orig_keypoint[p, 0], orig_keypoint[p, 1], f"{p}")
    plt.savefig(save_path)
    plt.close() 
    
def test_keypoints_plot(image, outputs, save_path):
    """
    This function plots the keypoints for the outputs and images
    from the test.
    """
    output_keypoints = outputs.reshape(NUM_KEYPOINTS, 2) * [IMG_ORIG_SIZE / IMGAE_RESIZE, IMG_ORIG_SIZE / IMGAE_RESIZE]
    image = image.reshape(IMG_ORIG_SIZE, IMG_ORIG_SIZE, 3)
    plt.style.use('default')
    plt.imshow(image)
    plt.axis('off')
    for p in range(output_keypoints.shape[0]):
        plt.plot(output_keypoints[p, 0], output_keypoints[p, 1], 'r.')
        plt.text(output_keypoints[p, 0], output_keypoints[p, 1], f"{p}")
    plt.savefig(save_path)
    plt.close()
    
def dataset_keypoints_plot(data):
    """
    This function shows the image faces and keypoint plots that the model
    will actually see. This is a good way to validate that our dataset is in
    fact correct and the faces align with the keypoint features. The plot 
    will be show if you execute `dataset.py`.
    """
    plt.figure(figsize=(20, 24))
    for i in range(30):
        img = data[0][0][i]
        img = np.array(img/255, dtype='float32')
        img = img.reshape(IMGAE_RESIZE, IMGAE_RESIZE, 3)
        plt.subplot(5, 6, i+1)
        plt.imshow(img)
        keypoints = data[0][1][i]
        keypoints = keypoints.reshape(NUM_KEYPOINTS, 2)
        for kp in keypoints:
            plt.plot(kp[0], kp[1], 'r.')
    plt.show()
    plt.close()
    
def save_plots(history, filename, skip=0):
    """
    Function to save the loss and accuracy plots to disk.
    """
    train_loss = history.history['loss'][skip:]
    valid_loss = history.history['val_loss'][skip:]
    # Loss plots.
    plt.figure(figsize=(12, 9))
    plt.plot(
        list(range(skip,len(train_loss)+skip)), train_loss, 
        color='orange', linestyle='-', label='train loss'
    )
    plt.plot(
        list(range(skip,len(valid_loss)+skip)), valid_loss, 
        color='red', linestyle='-', 
        label='validation loss'
    )
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig(os.path.join(
        '.', 'outputs', filename
    ))
    plt.show()
```

```python
import cv2
import pandas as pd
import numpy as np
from tensorflow.keras.utils import Sequence
from tqdm import tqdm
from sklearn.model_selection import train_test_split

from imgaug.augmentables.kps import KeypointsOnImage
from imgaug.augmentables.kps import Keypoint
import imgaug.augmenters as iaa

train_aug = iaa.Sequential(
    [
        iaa.Resize({"height": IMGAE_RESIZE, "width": IMGAE_RESIZE}),
        #iaa.Fliplr(0.3),
        #iaa.Sometimes(0.7, iaa.Affine(rotate=15, scale=(0.9, 1.1))),
#         iaa.Sometimes(0.3, iaa.Crop(px=(0, 24))), # crop images from each side by 0 to 16px (randomly chosen)
#         iaa.Sometimes(0.7, iaa.Affine(translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)})),
        #iaa.Sometimes(0.3, iaa.GaussianBlur(sigma=(0, 0.2))) # blur images with a sigma of 0 to 2.0
    ]
)

val_aug = iaa.Sequential([
        iaa.Resize({"height": IMGAE_RESIZE, "width": IMGAE_RESIZE})
    ])


def train_val_test_split(test_split, val_split=None):
    images = np.load('./images.npy')
    landmarks = np.load('./landmarks.npy')
    
    X, X_test, y, y_test = train_test_split(images, landmarks, test_size=test_split, random_state=SEED) 
    
    if MONITOR_VAL:
        X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=val_split, random_state=SEED) 
    
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    else:
        return X, X_test, y, y_test
    

class KeypointDataset(Sequence):
    def __init__(self, images, keypoints, batch_size, aug=None):
        self.batch_size = batch_size
        self.images = images
        self.keypoints = keypoints
        self.aug = aug
        
    def __len__(self):
        return len(self.images) // self.batch_size
    
    def __getitem__(self, index):
        batch_images = self.images[index*self.batch_size:(index+1)*self.batch_size]
        batch_keypoints = self.keypoints[index*self.batch_size:(index+1)*self.batch_size]
        final_images = []
        final_keypoints = []
        for j in range(self.batch_size):
            # Reshape the images into their original dimensions.
            input_image = batch_images[j]
            image = input_image.reshape(IMG_ORIG_SIZE, IMG_ORIG_SIZE, 3)
            
            # Get the keypoints.
            keypoints = batch_keypoints[j,]
            keypoints = np.array(keypoints, dtype='float32')
            # Reshape the keypoints.
            keypoints = keypoints.reshape(NUM_KEYPOINTS, 2)
                        
            if self.aug is not None:
                # We project the original image and its keypoint coordinates.
                kps_obj = KeypointsOnImage([],(0,0)).from_xy_array(keypoints, shape=image.shape)
                # Apply the augmentation pipeline.
                image, kps_obj = self.aug(image=image, keypoints=kps_obj)
                keypoints = kps_obj.to_xy_array()
                                
            keypoints = np.ravel(keypoints)
            
            final_images.append(image)
            final_keypoints.append(keypoints)
        final_images = np.array(final_images)
        final_keypoints = np.array(final_keypoints)
        return (final_images, final_keypoints)
```

```python
# Get the training and validation data samples.
if MONITOR_VAL:
    X_train, X_val, X_test, y_train, y_val, y_test = train_val_test_split(TEST_SPLIT, VAL_SPLIT)
else:
    X_train, X_test, y_train, y_test = train_val_test_split(TEST_SPLIT)

def get_data():
    
    if MONITOR_VAL:
        train_ds = KeypointDataset(X_train, y_train, batch_size=BATCH_SIZE, aug=train_aug)
        valid_ds = KeypointDataset(X_val, y_val, batch_size=BATCH_SIZE, aug=val_aug)
        test_ds = KeypointDataset(X_test, y_test, batch_size=BATCH_SIZE, aug=val_aug)

        return train_ds, valid_ds, test_ds
    
    else: 
        train_ds = KeypointDataset(X_train, y_train, batch_size=BATCH_SIZE)
        test_ds = KeypointDataset(X_test, y_test, batch_size=BATCH_SIZE)

        return train_ds, test_ds
```

```python
train_ds, val_ds, test_ds = get_data()
```
