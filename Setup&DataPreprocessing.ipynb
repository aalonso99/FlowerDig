```python
# Las dos funciones siguientes son para cargar las imágenes y los landmarks,
# respectivamente, la primera vez. Una vez se leen y se corrigen, como se verá 
# más adelante, las guardo con otro formato que permite leerlas mucho más rápido
# usando np.save y np.load.
# Ambas funciones leen las imágenes y landmarks de una carpeta. La función
# load_images_and_landmarks más adelante se encargará de recorrer todas las 
# carpetas y llamar a estas funciones para cada carpeta.

# El parámetro dim es el tamaño al que reducimos las imágenes.
def load_images_from_folder(folder, dim=(224,224), return_names=True):
    images = {}
    sizes = {}
    names = []  # Los nombres de cada imagen nos permitirán saber si cada imagen
                # tiene su lista de landmarks
    print(os.listdir(folder))

    for filename in os.listdir(folder):
        img = cv.imread(os.path.join(folder,filename))
        # Hay arhivos que no son imágenes, así que la línea anterior puede generar
        # img = None, en cuyo caso evitamos añadir la imagen a la lista
        if img is not None:     
            sizes[filename] = (img.shape[1], img.shape[0])
            images[filename] = cv.resize(img, dim)[:,:,::-1]
            names.append(filename.upper())

    if return_names:
        return images, sizes, names
    else:
        return images, sizes


# Esta función devuelve una lista formada por listas con los landmarks de todas las imágenes de una
# carpeta (todas las que aparecen en su archivo .TPS correspondiente).
# img_size es una dupla de la forma (x,y), ambos números enteros
# redim es el tamaño de la redimensión hecha por TPSDig de la imagen
# Un landmark es a su ver una lista con dos números, así que el resultado debería tener la forma:
#       landmarks = [[[x,y],...<-coordenadas]<-lista de landmarks de una imagen ... ] ... ] lista de listas de landmarks para todas las imágenes
def load_landmarks(folder, img_sizes, redim, all_redim=False, dim=224, return_names=True):
    landmarks = {}
    names = []
    landmarks_file_name = None

    for x in os.listdir(folder):
        if x.split('.')[-1].upper() == 'TPS':
            landmarks_file_name = x
            break

    if landmarks_file_name != None:
        with open(os.path.join(folder, landmarks_file_name)) as landmarks_file:
            aux_landmarks_list = None
            for line in landmarks_file:
                if line.split('=')[0] == 'LM':  # LM=32 marcar el principio de una imagen nueva...
                    aux_landmarks_list = []
                elif line[0].isdigit():         # Si una línea empieza con un número suponemos que...
                                                # se dan un par de coordenadas
                    aux_landmarks_list.append(list(map(int, map(float, line.split()))))  # Esto lee las coordenadas,...
                                                # las convierte a números enteros, las transforma en una lista...
                                                # formada por dos números y las añade a la cuenta
                elif line.split('=')[0] == 'IMAGE': # Las líneas que empiezan por IMAGE tienen el nombre de la 
                                                    # imagen a la que corresponden los landmarks anteriores
                    img_name = line.split('=')[1][:-1].upper()
                    names.append(img_name)   # El segundo elemento detrás del = es el nombre de la imagen seguido por
                                                            # un signo de salto de línea \n que no necesitamos.
                                                            # La indexación [:-1] sirve para eliminar el salto de línea.
                    landmarks[img_name] = aux_landmarks_list
    else:
        print("WARNING. Folder "+folder+" has no .tps file.")

    # Invertimos el eje vertical de cada imagen. El programa tpsDig empieza a contar desde abajo a la izquierda
    # y numpy y OpenCV cuentan desde arriba a la izquierda, lo que mostraba desajustes en las imágenes
    for img_name in names:
        img_lms = landmarks[img_name]
        for lm in img_lms:
            lm[1] = (img_sizes[img_name][1]-lm[1])

        # img_size/12 es una cota mínima arbitraria
        if all_redim or np.min(np.asarray(img_lms)[:,0]) < img_sizes[img_name][0]/6:
            for lm in img_lms:
                lm[0] = lm[0]/redim[0]*img_sizes[img_name][0]
                lm[1] = lm[1]/redim[1]*img_sizes[img_name][1]

        for lm in img_lms:
            lm[0] = int(lm[0]/img_sizes[img_name][0]*dim)
            lm[1] = int(lm[1]/img_sizes[img_name][1]*dim)
    
    
    if return_names:
        print('Names: '+str(len(names)))
        return landmarks, names
    else:
        return landmarks
            

# Esta función combina las dos anteriores, comprueba que cada imagen 
# cargada tiene una lista de landmarks asociados y solo las añade en ese caso.
# Recorremos 
def load_images_and_landmarks(dim=(224,224)):
    DIRS = [('T1', (770, 516), False), ('T2', (770, 516), False), 
            ('T3', (1735, 1299), True), ('T4', (1735, 1299), True), 
            ('T5', (1027, 686), True)]

    images = []
    landmarks = []
    current_dir = os.getcwd()
    for dir, redim, all_redim in DIRS:
        os.chdir(os.path.join(path,dir))
        SUBFOLDERS = [ x for x in os.listdir() if os.path.isdir(x) ]
        print(SUBFOLDERS)
        for subfolder in SUBFOLDERS:
            if subfolder[-1].isdigit():
                subfolder_images, img_sizes, img_names = load_images_from_folder(subfolder, dim)
                subfolder_landmarks, img_w_lms_names = load_landmarks(subfolder, img_sizes, redim, all_redim=all_redim)

                for img_name in img_w_lms_names:
                    images.append(subfolder_images[img_name])
                    landmarks.append(subfolder_landmarks[img_name])
        os.chdir(current_dir)

    return images, landmarks
```
